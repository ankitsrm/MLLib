```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.iloc[:, 0:4], # Select features from columns 0 to 3 (inclusive)
    df.iloc[:, -1], # Select the target variable from the last column
    test_size=0.2, # Allocate 20% of data for testing
    random_state=2 # Ensure reproducible results
)
```
